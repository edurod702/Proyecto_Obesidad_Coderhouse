## Contexto

<font size=4>Introducción

El sobrepeso y la obesidad se definen como una acumulación anormal o excesiva de grasa que puede ser perjudicial para la salud. Desde 1975, la obesidad se ha casi triplicado en todo el mundo. En 2016, más de 1900 millones de adultos de 18 o más años tenían sobrepeso, de los cuales, más de 650 millones eran obesos. Además, el 39% de las personas adultas de 18 o más años tenían sobrepeso, y el 13% eran obesas. Puede decirse que la mayoría de la población mundial vive en países donde el sobrepeso y la obesidad se cobran más vidas de personas que la insuficiencia ponderal. Si bien el sobrepeso y la obesidad se consideraban antes un problema propio de los países de ingresos altos, actualmente ambos trastornos aumentan en los países de ingresos bajos y medianos, en particular en los entornos urbanos. La causa fundamental del sobrepeso y la obesidad es un desequilibrio energético entre calorías consumidas y gastadas dado un aumento en la ingesta de alimentos de alto contenido calórico que son ricos en grasa acompañado de un descenso en la actividad física debido a la naturaleza cada vez más sedentaria de muchas formas de trabajo, los nuevos modos de transporte y la creciente urbanización. A menudo los cambios en los hábitos alimentarios y de actividad física son consecuencia de cambios ambientales y sociales asociados al desarrollo y de la falta de políticas de apoyo en sectores como la salud; la agricultura; el transporte; la planificación urbana; el medio ambiente; el procesamiento, distribución y comercialización de alimentos, y la educación. Determinar así factores determinantes en la obesidad es un aspecto crucial del sistema de salud, puesto que la obesidad podría prevenirse incidiendo sobre éstos.

<font size=4>Descripción del dataset

El dataset presenta datos para la estimación de los niveles de obesidad en individuos de los países de México, Perú y Colombia, con base en sus hábitos alimentarios y condición física. Los datos contienen 17 atributos y 2111 registros, los registros están etiquetados con la variable de clase NObesidad (Nivel de Obesidad), que permite clasificar los datos utilizando los valores de Peso Insuficiente, Peso Normal, Sobrepeso Nivel I, Sobrepeso Nivel II, Obesidad Tipo I, Obesidad Tipo II y Obesidad Tipo III. El 77% de los datos se generó de forma sintética utilizando la herramienta Weka y el filtro SMOTE, el 23% de los datos se recolectó directamente de los usuarios a través de una plataforma web.

## Hipótesis

El género, características físicas (edad, altura, peso), antecedentes familiares, hábitos de consumo alimenticio (control sobre ingesta calórica, comer verduras, número de comidas al día, toma de agua), vicios (fumar o beber), la actividad física (ejercicio, forma de desplazarse de un lugar a otro) y el tiempo de uso de dispositivos tecnológicos de forma sedentaria son factores determinantes en el sobrepeso y la obesidad de las personas.Un modelo predictivo del nivel de esta enfermedad (Obesity_Type_I, II, y III, Overweight_Level_I y II, su ausencia (Normal_Weight) y defecto (Insufficient_Weight)) puede establecerse así considerando dichas características.

## Objetivo

Emplear la información del dataset de obesidad para generar modelos de aprendizaje computacionales que permitan identificar el nivel de obesidad de un individuo y construir sistemas de recomendación que controlen los niveles de obesidad en la muestra estudiada.

```python
from seaborn.regression import regplot
def my_pairgrid(input_df):
    mpl.rcParams["axes.labelsize"] = 20 #Incrementa los tamaños de las etiquetas
    g = sns.PairGrid(input_df, vars=["Age", 	"Height", 	"Weight", 	"NCP", 	"CH2O", 	"FAF", "TUE"])
    
    "En la diagonal se hará un plot de tipo KDE"
    "kernel density estimate (KDE) es un método de visualización de la distribución de los datos"
    g = g.map_diag(sns.kdeplot, shade=True)
    "Para la matriz inferior, se mostrará un plot de regresión"
    g = g.map_lower(sns.regplot)
    "Para la matriz superior, se mostrará un plot de dispersión"
    g = g.map_upper(sns.scatterplot)
    return(g)
```

```python
# One-hot encoding de múltiples columnas
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from seaborn import load_dataset
import pandas as pd

transformer = make_column_transformer(
    (OneHotEncoder(), ['Gender','family_history_with_overweight', 'FAVC','SMOKE','SCC','MTRANS']),
    remainder='passthrough')

transformed = transformer.fit_transform(data)
transformed_data = pd.DataFrame(transformed, columns=transformer.get_feature_names())
```

```python
transformed_data
```

```python
# ----- Matriz de valores de características -----

X = transformed_data[['onehotencoder__x0_Female','onehotencoder__x0_Male','onehotencoder__x1_no',
                      'onehotencoder__x1_yes','onehotencoder__x2_no','onehotencoder__x2_yes','onehotencoder__x3_no',
                      'onehotencoder__x3_yes', 'onehotencoder__x4_no','onehotencoder__x4_yes','NCP','CAEC_encoded',
                      'CALC_encoded', 'Age', 'Height', 'Weight', 'FCVC', 'NCP', 'CH2O', 'FAF', 'TUE']] 
# ----- Vector de etiquetas -----
y = transformed_data['NObeyesdad']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101) # 80% training y 20% test

clf = DecisionTreeClassifier()  # Crear el objeto Decision Tree classifier 
clf = clf.fit(X_train,y_train)  # Entrenar el Decision Tree Classifier
y_pred = clf.predict(X_test)    # Predecir la respuesta para el conjunto de prueba
```

```python
# Métricas sobre DT classifier
from sklearn.metrics import confusion_matrix
import seaborn as sns
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score, recall_score

print(f"Accuracy: {round(accuracy_score(y_test,y_pred),3)}")
print(f"Recall: {round(recall_score(y_test,y_pred, average = 'macro'),5)}")
print(f"Precision: {round(precision_score(y_test,y_pred, average = 'macro'),3)}")
print(f"f1_score: {round(f1_score(y_test,y_pred, average = 'macro'),3)}")

target_labels = ['Obesity_Type_I','Obesity_Type_III','Obesity_Type_II','Overweight_Level_I','Overweight_Level_II',
                 'Normal_Weight','Insufficient_Weight']

plt.figure()
cm = confusion_matrix(y_test,y_pred)
s_cm = sns.heatmap(cm,cmap='plasma',annot=True, fmt='g',
            xticklabels=target_labels,
            yticklabels=target_labels)
s_cm.set(xlabel='Predicted',ylabel='Real')
plt.show()
```

```python
transformed_data['Age_norm'] = (transformed_data['Age']-transformed_data['Age'].min())/(transformed_data['Age'].max()-transformed_data['Age'].min())
transformed_data['Height_norm'] = (transformed_data['Height']-transformed_data['Height'].min())/(transformed_data['Height'].max()-transformed_data['Height'].min())
transformed_data['Weight_norm'] = (transformed_data['Weight']-transformed_data['Weight'].min())/(transformed_data['Weight'].max()-transformed_data['Weight'].min())
transformed_data['FCVC_norm'] = (transformed_data['FCVC']-transformed_data['FCVC'].min())/(transformed_data['FCVC'].max()-transformed_data['FCVC'].min())
transformed_data['NCP_norm'] = (transformed_data['NCP']-transformed_data['NCP'].min())/(transformed_data['NCP'].max()-transformed_data['NCP'].min())
transformed_data['CH2O_norm'] = (transformed_data['CH2O']-transformed_data['CH2O'].min())/(transformed_data['CH2O'].max()-transformed_data['CH2O'].min())
transformed_data['FAF_norm'] = (transformed_data['FAF']-transformed_data['FAF'].min())/(transformed_data['FAF'].max()-transformed_data['FAF'].min())
transformed_data['TUE_norm'] = (transformed_data['TUE']-transformed_data['TUE'].min())/(transformed_data['TUE'].max()-transformed_data['TUE'].min())
```

```python
# ----- Matriz de valores de características -----

X = transformed_data[['onehotencoder__x0_Female','onehotencoder__x0_Male','onehotencoder__x1_no',
                      'onehotencoder__x1_yes','onehotencoder__x2_no','onehotencoder__x2_yes','onehotencoder__x3_no',
                      'onehotencoder__x3_yes', 'onehotencoder__x4_no','onehotencoder__x4_yes','CAEC_encoded',
                      'CALC_encoded', 'Age_norm', 'Height_norm', 'Weight_norm', 'FCVC_norm', 'NCP_norm', 'CH2O_norm', 'FAF_norm', 'TUE_norm']] 
# ----- Vector de etiquetas -----
y = transformed_data['NObeyesdad']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101) # 80% training y 20% test

clf = DecisionTreeClassifier()  # Crear el objeto Decision Tree classifier 
clf = clf.fit(X_train,y_train)  # Entrenar el Decision Tree Classifier
y_pred = clf.predict(X_test)    # Predecir la respuesta para el conjunto de prueba
```

```python
# Métricas sobre DT classifier
from sklearn.metrics import confusion_matrix
import seaborn as sns
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score, recall_score

print(f"Accuracy: {round(accuracy_score(y_test,y_pred),3)}")
print(f"Recall: {round(recall_score(y_test,y_pred, average = 'macro'),5)}")
print(f"Precision: {round(precision_score(y_test,y_pred, average = 'macro'),3)}")
print(f"f1_score: {round(f1_score(y_test,y_pred, average = 'macro'),3)}")

target_labels = ['Obesity_Type_I','Obesity_Type_III','Obesity_Type_II','Overweight_Level_I','Overweight_Level_II',
                 'Normal_Weight','Insufficient_Weight']

plt.figure()
cm = confusion_matrix(y_test,y_pred)
s_cm = sns.heatmap(cm,cmap='plasma',annot=True, fmt='g',
            xticklabels=target_labels,
            yticklabels=target_labels)
s_cm.set(xlabel='Predicted',ylabel='Real')
plt.show()
```
